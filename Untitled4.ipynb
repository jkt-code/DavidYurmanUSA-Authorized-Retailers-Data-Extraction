```python
import pandas as pd 
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from bs4 import BeautifulSoup 
import time
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementNotInteractableException
import csv
```

```python
# Load CSV file 
df = pd.read_csv('C:/Users/User/usa_sample.csv') 
# Set up Selenium WebDriver (e.g., using Chrome) 
```

```python
def handle_popups(driver):
    try:
        # Wait for the cookies "Reject All" button and click it
        cookies_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'onetrust-reject-all-handler'))
        )
        cookies_button.click()
    except Exception as e:
        print("No cookies consent button found or could not click it:", e)

    try: 
        WebDriverWait(driver, 10)
        # Inject jQuery if it's not already present 
        if not driver.execute_script("return typeof jQuery !== 'undefined';"): 
            driver.execute_script(""" var jq = document.createElement('script'); 
            jq.src = 'https://ajax.googleapis.com/ajax/libs/jquery/3.5.1/jquery.min.js'; 
            document.getElementsByTagName('head')[0].appendChild(jq); """) 
            time.sleep(2) # Give jQuery time to load # Use jQuery to remove the signup modal 
        driver.execute_script("$('.modal, .modal-backdrop').remove();")
        print('Signup modal removed using jQuery')
        ("store-locator-wrapper")
    except (TimeoutException, NoSuchElementException) as e: 
        print("No signup modal close button found or could not click it:", e) 
        print(driver.page_source) # Print page source for debugging
    # Click the `a` element within the `div` and open its `href` link 
    try: 
        div_element = WebDriverWait(driver, 20).until( 
            EC.presence_of_element_located((By.CLASS_NAME, 'store-locator-wrapper')) 
        ) 
        anchor_element = div_element.find_element(By.CLASS_NAME, 'js-close-with-redirect') 
        href = anchor_element.get_attribute('href') 
        if href: 
            driver.get(href) 
            print(f'Navigated to {href}') 
        else: 
            print(f"No href found for anchor element") 
    except (TimeoutException, NoSuchElementException) as e: 
        print(f"Failed to find element with class 'js-close-with-redirect' or open href link:", e) 
        print(driver.page_source) # Print page source for debugging
    

# Function to scrape data using input values from CSV
def scrape_data(input_value):
    # Set up Selenium WebDriver (e.g., using Chrome)
    driver = webdriver.Chrome()  # Ensure you have the ChromeDriver installed and in PATH

    driver.get('https://www.davidyurman.com/stores?boutiques=true&retailers=true&radius=300')

    # Handle popups 
    handle_popups(driver)
    
    # Find the input element and enter the value
    input_element = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.NAME, 'postalCode'))
    )
    driver.execute_script("arguments[0].scrollIntoView(true);", input_element)
    input_element.clear()
    input_element.send_keys(input_value)
    
    # Use JavaScript executor to click the button
    search_button = driver.find_element(By.CSS_SELECTOR, 'button.btn.btn-primary.btn-storelocator-search')
    driver.execute_script("arguments[0].click();", search_button)
    # # Wait for the new page or element to load
    # WebDriverWait(driver, 20).until(
    #     EC.presence_of_element_located((By.CLASS_NAME, 'results-card'))  # Adjust class name as needed
    # )
    time.sleep(5)
    # Scrape data from the resulting page
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    results = soup.find('div', class_='results-card results-padding')  # Replace with the actual tag and class name of the data you want to scrape
    print('_'*50)
    print(input_value)
    print('_'*50)
    csv_filename = 'articles.csv' 
    with open(csv_filename, 'a', newline='') as file:
        writer = csv.writer(file) 
        if results:
            results1 = results.find('div', class_='store-locator__list even-count')
            if results1:
                results2 = results1.find_all('div',class_= re.compile('^store-locator__list__item'))
                for r in results2:
                    store_name = r.find('h3', class_='store-locator__list__item__title')
                    if store_name:
                        address_element = r.find('address', class_='store-text')
                        if address_element:
                            address_text = address_element.get_text(separator=' ', strip=True) 
                            # Extract the phone number dynamically 
                            phone_number_element = address_element.find('a', class_='storelocator-phone') 
                            if phone_number_element: 
                                phone_number = phone_number_element.get_text(strip=True) 
                                print("Address:", address_text.replace(phone_number, '').strip()) 
                                print("Phone Number:", phone_number)
                                writer.writerow([input_value,store_name.text,address_text.replace(phone_number, '').strip(),phone_number])
        else:
            print("No results")
    # Close the browser
    driver.quit()
    print('*'*50)

# Iterate over the CSV values and scrape data
for value in df['City']:
    scrape_data(value)
```
